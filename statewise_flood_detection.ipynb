```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.compose import ColumnTransformer
import pickle
```

```python
df = pd.read_csv('flood_risk_dataset_india.csv')
df1 = pd.read_csv('flood_risk_with_states_kd_mapped.csv')
```

```python
df.head()
```

```python
df1.head()
```

```python
df1 = df1['State Name']
```

```python
df = pd.concat([df1,df],axis=1)
```

```python
df.drop(['Latitude','Longitude'],axis=1,inplace=True)
```

```python
df
```

```python
df=df.sort_values(by ='State Name',ascending=True)
```

```python
charecter = [features for features in df.columns if df[features].dtype=='O']
for feat in charecter:
    print(feat,df[feat].unique(),'\n')
```

```python
X = df.drop('Flood Occurred',axis=1)
y= df['Flood Occurred']
```

```python
la = LabelEncoder()
```

```python
X['State Name']=la.fit_transform(X['State Name'])
```

```python
with open('state_emcoder.pkl','wb') as file1:
    pickle.dump(la,file1)
```

```python
x_train,x_test,y_train,y_test= train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
X_char = X.select_dtypes(include='object').columns
X_num = X.select_dtypes(exclude='object').columns
X_char
```

```python
Standardsclr = StandardScaler()
OneHotEncode= OneHotEncoder(drop='first',handle_unknown='ignore',sparse_output=False)
```

```python
Processor = ColumnTransformer([
    ('OneHotEncode',OneHotEncode,X_char),
    ('stand_sclr',Standardsclr,X_num)
    
])
```

```python
x_train = Processor.fit_transform(x_train)
x_test = Processor.transform(x_test)
```

```python
with open('Preprocess.pkl','wb') as file:
    pickle.dump(Processor,file)
```

```python
x_train.dtype
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasClassifier
```

```python
def create_model(neurons=32,layers=2):
    model= Sequential()
    model.add(Dense(neurons,activation='relu',input_shape=(x_train.shape[1],)))

    for _ in range(layers-1):
        model.add(Dense(neurons,activation='relu'))

    model.add(Dense(1,activation='sigmoid'))
    model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return model
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
model = KerasClassifier(layers=1,neurons=32,build_fn=create_model,verbose=1,callbacks=early_stop)
```

```python
parem_grids= {
    'neurons':[16,32,64,128],
    'layers':[1,2,3,4,],
    'epochs':[50,100,150,200,250]
}
```

```python
Grid_search = GridSearchCV(estimator=model,param_grid=parem_grids,n_jobs=-1,cv=3,verbose=1)
grid_result = Grid_search.fit(x_train,y_train)
```

```python
print('Best %f using %s',(grid_result.best_score_,grid_result.best_params_) )
```

```python
best_keras = grid_result.best_estimator_
best_keras_model = best_keras.model_
best_keras_model.save('Best_model.keras')
```
